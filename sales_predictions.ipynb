```python
low_fat = df.loc[df['Item_Fat_Content']=='Low Fat','Item_Weight']
regular = df.loc[df['Item_Fat_Content']=='Regular','Item_Weight']
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,8))
boxplots = axes.boxplot([low_fat, regular], 
            labels=['Low Fat', 'Regular'], 
            notch = True,
            widths = .5,
            patch_artist=True,
            medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
            boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
            );
boxplot1 = boxplots['boxes'][0]
boxplot1.set_facecolor('orange')
plt.ylabel('Fat content per item weight', fontsize = 20);
plt.xticks(fontsize = 16);
plt.yticks(fontsize = 16);
```
